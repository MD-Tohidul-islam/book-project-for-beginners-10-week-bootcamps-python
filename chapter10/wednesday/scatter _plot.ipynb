```python
# creating a scatter plot to represent height-weight distribution
from matplotlib import pyplot as plt
import random
```

```python
random.seed(2)
height = [randint(58,78) for x in range(20)] #20 records betweent 4'10" and 6'6"
weight = [randint(90,250) for x in range(20)] #20 records between 9lbs. and 250lbs
plt.scatter(weight,height)
plt.title('height-weight distribution')
plt.xlabel('weight(lbs)')
plt.ylabel("height (inches)")
plt.show( )
```
